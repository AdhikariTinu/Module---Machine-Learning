```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1yKlW0x5yyRQijmWbEg5YCsq4wqLl3_oB")
```

```python
df.head()
```

```python
df.info()
```

```python
df_3 = df.copy()
df_3 = df_3.dropna()
df_3 = df_3.drop_duplicates()
```

```python
df.info()
```

```python
def price_integer(item):
    if "₹" in str(item):
        item = str(item).replace('₹','')
    if "," in str(item):
        item = str(item).replace(',','')
    if "%" in str(item):
        item = str(item).replace('%','')

    return float(item)
```

```python
df_3['discounted_price'] = df_3['discounted_price'].apply(price_integer)
df_3['discounted_price'].head()
```

```python
df_3['actual_price'] = df_3['actual_price'].apply(price_integer)
df_3['actual_price'].head()
```

```python
df_3['rating_count'] = df_3['rating_count'].apply(price_integer)
df_3['rating_count'].head()
```

```python
df_3.head()
```

```python
df_3['discount_percentage'] = df_3['discount_percentage'].apply(price_integer)
df_3['discount_percentage'].head()
```

```python
df_3['rating'].unique()
```

```python
df_3 = df_3[~(df_3['rating']=='|')]
df_3.reset_index(drop=True)
df_3.head()
```

```python
df_3.info()
```

```python
df_3['rating'] = df_3['rating'].apply(price_integer)
df_3.info()
```

```python
df_3.describe()
```
